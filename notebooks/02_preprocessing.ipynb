## Data Preprocessing:

### Loading the data:

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer(as_frame=True)
df = data.frame
df.head(2)
```

#### Separation of Features and Label:

First, we separate input and output features in the data

```python
feature_cols = df.drop(columns=["target"])
target_cols = df["target"]
```

#### Train-test Split:

After that, we perform train-test split to separate training and test data.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature_cols, target_cols,  test_size=0.2, random_state=42)
```

```python
X_train.shape
```

#### Making of Column Transformer for standard scaling:

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

feature_columns = X_train.columns.tolist() 

preprocessor = ColumnTransformer(
    transformers=[
        ("scaler", StandardScaler(), feature_columns)
    ],remainder="passthrough"
)
```

#### Applying preprocessing transformer on training data: 

```python
X_train_scaled = preprocessor.fit_transform(X_train)
X_train_scaled
```
